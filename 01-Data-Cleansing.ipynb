## 1. Pré-processamento dos dados


### 1.1 Remoção de variáveis irrelevantes ao modelo

No conjunto de dados analisado, identificou-se a presença de variáveis potencialmente irrelevantes para o desenvolvimento do modelo preditivo, então foram retiradas: 

- **id:** Identificação única atribuída pela LC ao empréstimo;
- **member_id:** Identificação única atribuída pela LC ao mutuário;
- **url:** URL para a página da LC com dados de listagem;
- **desc:** Descrição do empréstimo fornecida pelo mutuário;
- **title:** O título do empréstimo fornecido pelo mutuário;
- **zipcode:** Os 3 primeiros números do código postal fornecidos pelo mutuário no pedido de empréstimo;
- **emp_title:** O cargo fornecido pelo mutuário ao solicitar o empréstimo;
- **pymnt_plan:** Indica se um plano de pagamento foi estabelecido para o empréstimo;
- **application_type:** "Indica se o empréstimo é uma aplicação individual ou uma aplicação conjunta com dois co-mutuário.

Adicionalmente, a variável `sub_grade` foi eliminada do conjunto de dados devido à sua redundância com a variável `grade`:

- **sub_grade:** Subgrau de empréstimo atribuído pela LC;
- **grade:** Grau de empréstimo atribuído pela LC.

Por fim, optou-se por descartar as variáveis  `collection_recovery_fee` e `total_rec_late_fee`, por serem dados sobre o futuro.

- **collection_recovery_fee:** Cobrança de taxa de coleta pelo correio; 
- **total_rec_late_fee:** Taxas atrasadas recebidas até o momento;
- **next_pymnt_d:** Próxima data de pagamento agendada.

```python
raw_data.drop(['id', 'member_id', 'emp_title', 'url', 'desc', 'title', 'zip_code', 'sub_grade', 
               'collection_recovery_fee', 'total_rec_late_fee', 'policy_code', 'pymnt_plan', 
               'application_type', 'next_pymnt_d'], inplace = True, axis = 1)
```

### 1.2 Criação da variável dependente

A variável `loan_status`, que descreve a situação atual do empréstimo, possui várias classes e foi utilizada para a criação da variável resposta.

```python
# classes únicas da variável loan_status

raw_data.loan_status.unique()
```

```python
plt.title('Distribuição da Variável Status do Empréstimo')
sns.countplot(y = raw_data.loan_status);
```

```python
# cálculo do número de observações para cada classe da variável loan_status

raw_data['loan_status'].value_counts()
```

```python
# proporção de observações para cada classe da variável

raw_data['loan_status'].value_counts() / raw_data['loan_status'].count()
```

Devido a série de valores atribuídos a variável `loan_status`, optou-se por combinar seus valores em apenas dois, sendo que 0 irá representar os clientes "ruins", pessoas que entraram em default, e o 1 irá representar os clientes "bons", aqueles que não entraram em default.

```python
# criando a variável target 

raw_data['target'] = np.where(raw_data['loan_status'].isin(['Charged Off', 'Default',
                                                            'Does not meet the credit policy. Status:Charged Off',
                                                            'Late (31-120 days)']), 0, 1)
```

É perceptível o desbalanceamento das classes, o que é comum em datasets de casos de inadimplência ou fraude. 

```python
plt.title('Distribuição da Variável Target')
sns.countplot(x = raw_data.target, palette = ['red', 'green']);
```

```python
# percentual de classe da variável target criada

raw_data.target.value_counts(normalize=True) * 100
```

### 1.3 Tratamento dos dados

#### 1.3.1 Padronização de variáveis

Primeiramente foi realizada a transformação de algumas variáveis do tipo string para o tipo inteiro, sendo elas: 
- **emp_length:** Tempo de emprego em anos. Os valores possíveis estão entre 0 e 10, onde 0 significa menos de um ano e 10 significa dez ou mais anos;
- **term:** O número de pagamentos sobre o empréstimo. Os valores são em meses e podem ser de 36 ou 60.

```python
# classes únicas da variável emp_length

raw_data.emp_length.unique()
```

```python
labels = [ '< 1 year','1 year','2 years','3 years','4 years','5 years','6 years','7 years','8 years','9 years','10+ years']

plt.figure(figsize=(12,4))
plt.title('Target vs Tempo de Empregado')
sns.countplot(x = 'emp_length', data = raw_data, order = labels, hue = 'target', palette = ['red', 'green']);
```

```python
# substituindo strings

raw_data['emp_length'] = raw_data['emp_length'].str.replace('< 1 year', str(0), regex = True)
raw_data['emp_length'] = raw_data['emp_length'].str.replace('\+ years', '', regex = True)
raw_data['emp_length'] = raw_data['emp_length'].str.replace(' years', '', regex = True)
raw_data['emp_length'] = raw_data['emp_length'].str.replace(' year', '', regex = True)
raw_data['emp_length'] = raw_data['emp_length'].replace(np.nan, 0, regex = True)


# transformando dados de string para inteiro

raw_data['emp_length'] = pd.to_numeric(raw_data['emp_length'])
```

```python
raw_data.emp_length.unique()
```

```python
raw_data.term.unique()
```

Empréstimos com 36 meses são mais aprovados e/ou solicitados do que os com 60 meses.

```python
plt.figure(figsize=(12,4))
plt.title('Target vs Parcelas Empréstimo')
sns.countplot(x = 'term', data = raw_data, hue = 'target', palette = ['red', 'green']);
```

```python
# substituindo strings

raw_data['term'] = raw_data['term'].str.replace(' months', '', regex = True)
raw_data['term'] = raw_data['term'].replace(np.nan, 0, regex = True)


# transformando dados de string para inteiro

raw_data['term'] = pd.to_numeric(raw_data['term'])
```

#### 1.3.2 Tratamento de dados faltantes

Foram realizados diferentes tratamentos para os dados faltantes que serão descritos a seguir.

Para a variável `annual_inc` foi-se utilizada a média.

- **annual_inc:** A renda anual declarada pelo devedor do empréstimo durante o registro.

```python
raw_data['annual_inc'].fillna(raw_data['annual_inc'].mean(), inplace = True)
```

`total_rev_hi_lim` o total rotativo crédito/limite de crédito possui alguns valores faltando, então faz sentido que os valores ausentes sejam iguais a `funded_amnt`, O valor total comprometido com esse empréstimo naquele momento.

```python
raw_data['total_rev_hi_lim'].fillna(raw_data['funded_amnt'], inplace = True)
```

Remoção de colunas com mais de 90% dos valores nulos.

```python
porcentagem_nulos = raw_data.isnull().mean() * 100
colunas_a_manter = porcentagem_nulos < 90
raw_data = raw_data.loc[:, colunas_a_manter]
```

#### 1.3.3 Criação de dummies

Foram criadas variáveis dummies das seguintes variáveis categóricas: `grade`, `home_ownership`, `verification_status`, `purpose`, `initial_list_status`

```python
var_dummies = [pd.get_dummies(raw_data['grade'], prefix = 'grade', prefix_sep = ':'),
               pd.get_dummies(raw_data['home_ownership'], prefix = 'home_ownership', prefix_sep = ':'),
               pd.get_dummies(raw_data['verification_status'], prefix = 'verification_status', prefix_sep = ':'),
               pd.get_dummies(raw_data['purpose'], prefix = 'purpose', prefix_sep = ':'),
               pd.get_dummies(raw_data['addr_state'], prefix = 'addr_state', prefix_sep = ':'),
               pd.get_dummies(raw_data['initial_list_status'], prefix = 'initial_list_status', prefix_sep = ':')]

var_dummies = pd.concat(var_dummies, axis = 1)
raw_data = pd.concat([raw_data, var_dummies], axis = 1)
```

Retirando símbolo % de `int_rate` e `revol_util`

- **int_rate:** Taxa de juros sobre o empréstimo;
- **revol_util:** Taxa de utilização da linha rotativa, ou a quantidade de crédito que o mutuário está usando em relação a todo crédito rotativo disponível.

```python
# retirando o símbolo de porcentagem

raw_data['int_rate'] = raw_data['int_rate'].replace('%', '', regex = True)


# convertendo int_rate para número

raw_data['int_rate'] = pd.to_numeric(raw_data['int_rate'])
```

```python
sns.kdeplot(raw_data['int_rate'], hue=raw_data['target']);
```

```python
# retirando o símbolo de porcentagem

raw_data['revol_util'] = raw_data['revol_util'].replace('%', '', regex = True)


# convertendo revol_util para numérico

raw_data['revol_util'] = pd.to_numeric(raw_data['revol_util'])
```
