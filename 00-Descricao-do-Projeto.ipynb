# 1. Modelagem da Probabilidade de Risco de Crédito e Cálculo da Perda Esperada
---

No atual cenário econômico, a concessão de crédito emerge como uma decisão revestida de incerteza, configurando-se como um desafio substancial. Esta realidade aplica-se a uma variedade de contextos financeiros, incluindo empréstimos, financiamentos e transações comerciais a prazo. Inerente a essas operações está a potencialidade de ocorrência de perdas. Apesar da implementação de critérios rigorosos para a concessão de crédito, que abarcam a análise da saúde financeira do solicitante, a qualidade do relacionamento com o credor e o histórico de pagamentos prévios, persiste o risco de inadimplemento por parte do requerente. Tal inadimplemento acarreta a não recuperação do capital investido e dos juros correspondentes, culminando em interrupções nos fluxos de caixa e escalada nos custos associados à cobrança de dívidas (SICSÚ, 2010).

Contudo, a habilidade de estimar a probabilidade de risco de uma operação de crédito e a identificação prévia de solicitantes potencialmente inadimplentes podem conferir maior segurança à decisão de concessão de crédito, mitigando, assim, as perdas institucionais. Neste contexto, o objetivo principal deste projeto consistiu na modelagem de risco de crédito. Para tanto, foram empregadas metodologias de aprendizado de máquina e técnicas estatísticas avançadas, visando à identificação de variáveis preditoras de inadimplência e à estimativa da **Perda Esperada**.

Ressalta-se a importância de se aprofundar no entendimento do ciclo de crédito, o qual, apesar de apresentar variações conforme o modelo de negócio em questão, geralmente compreende as seguintes etapas.

## 1.1 Ciclo de Crédito
---

O ciclo de crédito constitui um componente fundamental nas estratégias operacionais de entidades que ofertam essa modalidade de benefício. Este processo é imprescindível para a minimização de riscos e perdas financeiras, bem como para a otimização das atividades relacionadas à concessão de crédito. Adotando essa abordagem, as empresas podem antever potenciais inadimplências ou efetuar a recuperação de ativos financeiros de maneira eficaz (ZUÑIGA, 2019).

Este ciclo é composto pelas etapas descritas a seguir:

<img src = "https://images2.imgbox.com/23/f6/vHEZbiCf_o.png" width="450">

-  **Prospecção:** Neste estágio inicial, a entidade delineia o perfil de cliente ideal, caracterizado pela confiabilidade e pontualidade no cumprimento de suas obrigações financeiras. Tal perfil é construído mediante a aplicação de filtros analíticos sobre o conjunto de dados disponíveis acerca dos clientes; 
-  **Concessão de crédito:** Nesta fase, ocorre a avaliação do risco associado à concessão de crédito para cada cliente. Esta análise é realizada à luz do perfil ideal previamente estabelecido, bem como das políticas internas da empresa concernentes à concessão de crédito; 
-  **Gestão de risco:** Aqui, a empresa empreende um monitoramento contínuo dos clientes e de suas respectivas situações financeiras, com o intuito de prevenir o surgimento de casos de inadimplência e assegurar a manutenção de um relacionamento mutuamente benéfico entre cliente e empresa; 
-  **Cobrança:** Embora as fases anteriores visem à redução do risco de inadimplência, o risco residual permanece. Na eventualidade de inadimplência, esta etapa contempla a implementação de estratégias de negociação, preditas pelo sucesso das etapas precedentes na mitigação de riscos. 

<br>


Apesar de as medidas preventivas e as políticas de gestão de risco contribuírem significativamente para a redução da inadimplência, a concessão de crédito a prazo representa um desafio complexo para qualquer natureza de negócio. Diante deste panorama, a estratégia ótima consiste no controle rigoroso e na redução dos riscos associados à operação, por meio de uma análise abrangente dos solicitantes de crédito, isto é, mediante a avaliação detalhada do risco de crédito. (NEOWAY, 2020).

## 1.2 Risco de Crédito
---

O **Risco de Crédito** é conceituado como a probabilidade de um devedor falhar no cumprimento de suas obrigações financeiras, decorrentes de acordos de parcelamento, empréstimos, financiamentos, entre outros, perante o credor. Este risco, atribuído a cada cliente, pode ser originado tanto de características intrínsecas ao cliente quanto do contexto econômico, ou de uma intersecção de ambos. Estes fatores são meticulosamente avaliados no processo de cálculo de risco (GEORGIEV, 2022).

Neste panorama, os modelos de **credit scoring** (pontuação de crédito) desempenham um papel crucial na concessão de crédito, auxiliando na determinação da viabilidade de uma operação e na potencialidade de geração de perdas para o credor, visando à mitigação do **risco de crédito**.

A precisão na estimativa do risco torna-se fundamental, com o objetivo de reduzir ao mínimo a **Perda Esperada**. Este cálculo considera a probabilidade de inadimplência, a taxa de perda e o volume de exposição ao risco.

Os componentes são definidos como:

- **Probabilidade de Inadimplência (Probability of Default - PD):** É a probabilidade de um devedor falhar no pagamento das parcelas devidas;
- **Índice de Perda por Inadimplência (Loss Given Default - LGD):** Representa o percentual de perda financeira para o credor em caso de inadimplência do devedor;
- **Exposição em Caso de Inadimplência (Exposure at Default - EAD):** Refere-se ao montante total que o credor deixa de recuperar em razão da inadimplência do devedor.

A formulação para o cálculo da **Perda Esperada** é expressa por:

`Perda esperada` = `Probabilidade de Inadimplência (%)` x `Índice de Perda (%)` x `Exposição (R$)`

---

Para facilitar a compreensão, consideremos o exemplo a seguir (*desconsiderando-se taxas de juros*):

Imagine-se que um indivíduo solicita a uma instituição financeira um empréstimo de R\\$ 400.000,00 para a aquisição de um imóvel. Inicialmente, realiza-se uma análise das informações pessoais do solicitante para determinar sua pontuação de crédito, isto é, a probabilidade de que o solicitante não efetue o pagamento do valor requisitado, acrescido de juros.

Supondo que o solicitante efetue um pagamento inicial de R\\$ 40.000,00 ao banco, restando um saldo devedor de R\\$ 360.000,00, e, subsequente a isso, ocorra um default, a **Exposição em Caso de Inadimplência** corresponderá ao valor remanescente não pago.

Contudo, caso o banco consiga vender o imóvel por R\\$ 342.000,00, a **Exposição em Caso de Inadimplência** será ajustada para R\\$ 18.000,00 (R\\$ 360.000,00 - R\\$ 342.000,00), resultando em um **Índice de Perda por Inadimplência** de 5% (R\\$ 18.000,00 ÷ R\\$ 360.000,00).

Considerando que um em cada quatro solicitantes na instituição resulta em default, a **Probabilidade de Inadimplência** seria de 25%.

Portanto, a **Perda Esperada** seria calculada como: `25% x 5% x R$ 360.000,00 = R$ 4.500,00`.

# 2. Problema 
---

Diversas metodologias são empregadas na estimativa da perda esperada, cada uma apresentando diferentes graus de complexidade e precisão. A escolha da técnica apropriada exige das instituições financeiras uma avaliação cuidadosa de suas prioridades, recursos disponíveis e objetivos estratégicos específicos. No contexto deste projeto, optou-se pela aplicação de métodos estatísticos avançados, nomeadamente a regressão logística e a regressão linear, como ferramentas preditivas centrais.

A **regressão logística** foi selecionada para a previsão da probabilidade de inadimplência devido à sua capacidade de lidar com variáveis dependentes binárias – neste caso, a ocorrência ou não de inadimplência. Esse método permite estimar a probabilidade de um evento ocorrer, baseando-se em um conjunto de variáveis independentes. Isso é particularmente útil para identificar os fatores que influenciam a probabilidade de inadimplência de um devedor, permitindo à instituição tomar decisões mais informadas no que tange à concessão de crédito.

Por outro lado, a **regressão linear** foi utilizada para calcular o índice de perda e a exposição ao risco em caso de inadimplência. Este método é adequado para modelar a relação entre uma variável dependente contínua e uma ou mais variáveis independentes. No âmbito do projeto, isso implica na determinação de como diferentes fatores contribuem para o valor da perda e da exposição ao risco, proporcionando insights valiosos para a mitigação de perdas potenciais.

O cerne deste projeto visa responder à questão fundamental: *Como modelar a probabilidade de risco de crédito e calcular a perda esperada?* Para abordar esta questão, foi essencial não apenas selecionar métodos estatísticos apropriados, mas também implementá-los de forma a capturar as nuances do comportamento de inadimplência e os impactos financeiros associados. Isso envolveu a construção de modelos preditivos que são tanto robustos quanto capazes de fornecer estimativas precisas de risco e perda, fundamentais para a tomada de decisão no âmbito do crédito.

Além disso, este projeto se propôs a explorar as interações entre diferentes variáveis econômicas e pessoais dos solicitantes de crédito, analisando como essas interações afetam a probabilidade de inadimplência e, consequentemente, a perda esperada pela instituição. A abordagem adotada procurou, portanto, não apenas fornecer estimativas quantitativas, mas também oferecer insights qualitativos que poderiam guiar estratégias de mitigação de risco mais eficazes e políticas de concessão de crédito mais assertivas.

# 3. Impacto
---

O Brasil se destaca por apresentar significativos índices de inadimplência, um desafio que persiste e se intensifica no cenário econômico atual. De acordo com dados divulgados pelo Serasa em junho de 2022, através do [Mapa da Inadimplência](https://www.serasa.com.br/assets/cms/2022/MKTECS-965-Mapa-da-inadimplencia-JUNHO.pdf), observou-se um incremento de 200 mil novos casos de inadimplência no decorrer do ano, elevando o total para 66,8 milhões de brasileiros com restrições de crédito (SERASA, 2022). Este cenário ressalta a urgência e a relevância de abordagens inovadoras na gestão de crédito.

A implementação de soluções eficazes para a categorização de solicitantes de crédito é essencial, não apenas para identificar indivíduos com alta probabilidade de inadimplência, mas também para refinar as políticas de concessão de crédito, evitando a exclusão de potenciais bons pagadores por critérios excessivamente restritivos. Tais soluções têm o potencial de reduzir significativamente as perdas financeiras enfrentadas por diversas empresas, além de contribuir para o aprimoramento das metodologias empregadas na avaliação do risco de crédito e no cálculo da perda esperada.

Neste contexto, a necessidade de conduzir estudos focados na análise da inadimplência e na otimização das práticas de concessão de crédito no Brasil é evidente. Portanto, o desenvolvimento de trabalhos que visem à compreensão aprofundada e à inovação na gestão de risco de crédito emerge como fundamental para o avanço do conhecimento na área e para a formulação de estratégias mais eficazes e inclusivas de concessão de crédito.

O impacto potencial de avanços nesta área estende-se além da esfera financeira, promovendo a inclusão financeira e o desenvolvimento econômico ao possibilitar o acesso ao crédito a um espectro mais amplo da população. Assim, investir em pesquisa e desenvolvimento de novas metodologias para a avaliação de risco de crédito não apenas beneficia o setor financeiro, mas também contribui de maneira significativa para a sociedade como um todo, ao facilitar condições mais justas e acessíveis de financiamento.


# 4. Desenvolvimento da Solução
---

O desenvolvimento desta solução inovadora iniciou com a aplicação de uma metodologia criteriosa para a seleção de variáveis (features) e a construção de um conjunto de dados (dataset) robusto, empregando técnicas estatísticas avançadas. Esta fase inicial envolveu a seleção cuidadosa de um dataset representativo, seguida de uma análise detalhada das variáveis, utilizando métricas como o peso da evidência (Weight of Evidence - WoE) e o valor de informação (Information Value - IV). Após esta etapa de seleção, os dados foram organizados em categorias através das técnicas de coarse e fine classing, visando uma segmentação eficiente e significativa das variáveis.

A etapa subsequente consistiu em uma análise exploratória dos dados, com o objetivo de examinar a correlação e distribuição das variáveis selecionadas. Esta análise foi fundamental para obter insights valiosos sobre a estrutura dos dados, facilitando assim a modelagem subsequente. Com base nesses insights, procedeu-se ao desenvolvimento de um modelo de probabilidade de inadimplência empregando regressão logística, uma técnica que permite estimar a probabilidade de ocorrência de um evento, neste caso, a inadimplência, com base nas características dos mutuários.

O passo seguinte envolveu a elaboração de um sistema de pontuação de crédito (credit scoring), derivado dos coeficientes obtidos no modelo de regressão logística. Esta etapa permitiu a classificação dos mutuários de acordo com o risco de inadimplência, possibilitando a aplicação prática do modelo em decisões de concessão de crédito.

Para assegurar a eficácia e a estabilidade do modelo ao longo do tempo, implementou-se um mecanismo de monitoramento baseado no Índice de Estabilidade Populacional (Population Stability Index - PSI). Além disso, realizou-se uma análise comparativa entre os dados utilizados no desenvolvimento do modelo e um novo conjunto de dados, visando validar a robustez e a generalidade do modelo proposto.

Paralelamente, modelos preditivos para o cálculo do índice de perda e da exposição ao risco em casos de inadimplência foram desenvolvidos utilizando regressão linear, complementando a análise de risco de crédito. Finalmente, a perda esperada foi calculada integrando os resultados dos três modelos: probabilidade de inadimplência, índice de perda e exposição ao risco, oferecendo uma ferramenta abrangente e precisa para a gestão do risco de crédito.

# 5. Dados
---

O conjunto de dados empregado neste projeto oferece uma visão abrangente e detalhada do mercado de empréstimos ao consumidor, cobrindo um período extenso de treze anos, de 2007 a 2020. Originário do Lending Club, uma das maiores plataformas de empréstimo entre pares (peer-to-peer lending) nos Estados Unidos, este dataset engloba informações relativas a mais de 2 milhões de empréstimos concedidos ao longo deste intervalo temporal. 

A riqueza de dados incluída neste conjunto proporciona uma base sólida para análises profundas e modelagens preditivas no contexto de crédito ao consumidor. O Lending Club se destaca por facilitar empréstimos diretamente entre indivíduos, sem a intermediação de instituições financeiras tradicionais, o que confere características únicas aos dados coletados. 

Diversas versões deste conjunto de dados estão disponíveis, refletindo atualizações e expansões ao longo do tempo. Para os propósitos deste projeto, a versão específica utilizada foi retirada da plataforma Kaggle, sob o título "Lending Club Loan Data", e pode ser acessada através do seguinte link: [Lending Club Loan Data](https://www.kaggle.com/datasets/ethon0426/lending-club-20072020q1). Esta versão foi escolhida por sua abrangência e pela qualidade dos dados, facilitando assim a condução de análises precisas e o desenvolvimento de modelos estatísticos robustos.

O acesso a um dataset tão extenso e detalhado permite não apenas a modelagem de riscos de crédito e a previsão de inadimplências com alta precisão, mas também oferece insights valiosos sobre as tendências de empréstimos ao consumidor, padrões de pagamento, e outros fatores cruciais no âmbito de finanças pessoais e comportamento de crédito.

# 6. Resultados 
---

Este segmento apresenta uma análise detalhada dos resultados obtidos a partir dos modelos desenvolvidos ao longo deste projeto, enfocando tanto a precisão das previsões quanto as implicações práticas desses resultados.

- **Modelo de Probabilidade de Inadimplência:**

   - O modelo de regressão logística alcançou uma acurácia de 59% e uma pontuação ROC-AUC de 70%, evidenciando sua capacidade de prever o risco de crédito de maneira eficaz sem indicar overfitting. Tal performance é reforçada pelo F-1 score de 71% e pelo Recall de 57%, que atestam a habilidade do modelo em equilibrar precisão e sensibilidade.
   
   - As métricas de KS (Kolmogorov-Smirnov) e Gini, com valores de 0.30 e 0.38, respectivamente, sugerem que o modelo possui uma capacidade razoavelmente boa de diferenciar entre os mutuários que irão ou não entrar em inadimplência.
   

- **Modelo de Índice de Perda por Inadimplência:** (CORRIGIR)

    - A análise utilizando regressão logística revelou uma acurácia de 62% e ROC-AUC de 66%, o que indica uma capacidade competente do modelo em prever a taxa de recuperação dos valores em caso de inadimplência, sem indicar overfitting. F-1 score e Recall foram reportados como 60% e 71%, respectivamente.
    
    - Por outro lado, a abordagem de regressão linear mostrou um valor RMSE de 0.10, sugerindo um desvio padrão de previsão relativamente baixo, o que evidencia a precisão do modelo nesta dimensão de análise.


- **Modelo de Exposição em Caso de Inadimplência:** (CORRIGIR)

    - A análise dos resultados demonstrou uma distribuição de dados concentrada próxima a zero, o que é metodologicamente aceitável e alinha-se com as expectativas baseadas nos dados analisados. A correlação identificada, embora baixa, é considerada satisfatória para a aplicação do modelo EAD, indicando que, caso fosse significativamente mais alta, seria necessária uma revisão do modelo.


- **Cálculo de Perda Esperada:** (CORRIGIR)

    - A aplicação dos modelos anteriores ao dataset selecionado revelou que a perda esperada média é de U\\$ 592.28, com um desvio padrão de U\\$ 1.639.46, e um valor máximo de perda de U\\$ 37.975,99. A soma total da perda esperada para o portfólio foi calculada em U\\$ 519.819.141,88, representando uma proporção de 3.8% do valor total dos empréstimos.


Estes resultados são fundamentais para a compreensão e gestão do risco de crédito, oferecendo insights valiosos para a otimização de políticas de concessão de crédito e estratégias de recuperação de valores. Para projetos futuros, essas descobertas incentivam a exploração de novas técnicas e abordagens analíticas, com o objetivo de aprimorar ainda mais a precisão dos modelos e a eficiência das práticas de gestão de risco de crédito.    


# 7. Conclusão e Recomendações para Projetos Futuros
---

Este projeto alcançou êxito na concepção e implementação de um modelo preditivo para a probabilidade de inadimplência utilizando regressão logística, além de desenvolver dois modelos complementares com regressão linear para calcular o índice de perda por inadimplência e a exposição em casos de inadimplência. A integração desses modelos no cálculo da perda esperada cumpriu com êxito os objetivos propostos, demonstrando eficácia e relevância em seu contexto de aplicação.

Com base nos resultados obtidos e nas lições aprendidas, propõe-se para iniciativas futuras a elaboração de um modelo de behavior score. Esse modelo visa aprofundar a análise sobre o comportamento dos mutuários, considerando o histórico detalhado de suas transações. O objetivo é refinar a capacidade de classificação dos solicitantes de crédito, ajustando de maneira mais dinâmica os critérios e limites de concessão ao longo do tempo, em resposta às mudanças nos padrões de comportamento dos consumidores.

Além disso, é fundamental o monitoramento contínuo do modelo de probabilidade de inadimplência. Tal acompanhamento permite a detecção de mudanças no perfil dos solicitantes de crédito e na dinâmica do mercado, indicando quando poderia ser necessário desenvolver e implementar uma versão atualizada do modelo para manter sua precisão e efetividade.

Outra recomendação estratégica envolve a exploração de novas técnicas de seleção de variáveis e a aplicação de metodologias inovadoras na construção de modelos preditivos. Esse esforço tem o potencial de descobrir abordagens mais eficientes e econômicas que não somente aprimorem a performance dos modelos existentes, mas também contribuam para a redução de custos operacionais e de implementação. Essa frente de trabalho promete abrir caminhos para a otimização de processos e para a obtenção de vantagens competitivas significativas no setor de crédito.

Em síntese, as experiências e aprendizados decorrentes deste projeto pavimentam o caminho para futuras investigações e desenvolvimentos no campo da gestão de risco de crédito. A contínua busca por inovação e aprimoramento não só contribuirá para a sustentabilidade e competitividade das instituições financeiras, mas também para a promoção de práticas de concessão de crédito mais justas, transparentes e inclusivas.
